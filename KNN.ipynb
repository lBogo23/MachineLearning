```python
# Utilizar sklearn para implementar o KNN pois ja tem uma implementação da mesma
#Shift + enter = execução célula
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
```

```python
# Declarar uma variável para KNeighborsClassifier = (parametro)
#O parametro, conforme o código, para informar ao knn que cada novo elemento será usado para comparar
#Modelo KNN
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
data = pd.read_table('fruit_data_with_colors1.txt')
```

```python
data
```

```python
# Passar atributos dentro da lista
X = data[['mass','height','width','color_score']]
y = data['fruit_label']
```

```python
X
```

```python
y
```

```python
#Divisão de treino e teste
from sklearn.model_selection import train_test_split
```

```python
#Normalmente se usa X (maisculo) em aprendizado máquina
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
# Armazena os valores de treino no knn
knn.fit(X_train,y_train)
```

```python
# É aleatório
knn.score(X_test,y_test)
```

```python
# A ordem de grandeza dos dados faz diferença
#Para balançar os dados trabalhamos com escala [MinMaxScaler] (escalar os dados)
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
```

```python
# Iremos executar de maneira separada a transformação de escala
#Esse comando está pegando so dados de treino e transformando entre 0 e 1
X_train = mm.fit_transform(X_train)
```

```python
X_train
```

```python
# Estamos apenas trandformando nossos dados de acordo com a escala dos dados de treino
X_test = mm.transform(X_test)
```
